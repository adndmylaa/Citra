```python
# Import Library
from PIL import Image
import math
import matplotlib.pyplot as plt
```

```python
# Fungsi untuk mengubah pixel RGB menjadi grayscale
def convert_to_grayscale(pixel):
    red, green, blue = pixel
    gray_value = int(0.299 * red + 0.587 * green + 0.114 * blue)
    return gray_value
```

```python
# Fungsi untuk mengubah nilai grayscale menjadi biner
def convert_to_binary(gray_value, threshold=230):
    if gray_value >= threshold:
        return 1
    else:
        return 0
```

```python
# Membuka gambar dan mengonversinya ke mode RGB
gambar = Image.open('lukisan.jpeg')
gambar = gambar.convert('RGB')
width, height = gambar.size
```

```python
# Membuat matriks kosong untuk menyimpan nilai grayscale
binary_matrix = []
```

```python
# Proses konversi ke biner
for y in range(height):
    row = []
    for x in range(width):
        pixel = gambar.getpixel((x, y))
        grayscale_value = convert_to_grayscale(pixel)
        binary_value = convert_to_binary(grayscale_value, threshold=100)
        row.append(binary_value)
    binary_matrix.append(row)
```

```python
# Buat gambar baru dari hasil konversi biner
binary_image = Image.new('1', (width, height))
```

```python
# Isi gambar dengan nilai biner
for y in range(height):
    for x in range(width):
        color = binary_matrix[y][x]
        binary_image.putpixel((x, y), color)
```

```python
# Menampilkan gambar asli
plt.imshow(gambar)
plt.axis('off')
plt.title('Gambar Asli')
plt.show()
```

```python
# Tampilkan gambar hasil konversi biner
plt.imshow(binary_image)
plt.axis('off')
plt.title('Hasil Konversi Biner')
plt.show()
```
