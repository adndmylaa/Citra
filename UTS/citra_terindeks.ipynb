```python
from PIL import Image
import math
import matplotlib.pyplot as plt
```

```python
# mendefinisikan pallete warna
palette = {
    "Hitam": (0, 0, 0),
    "Merah": (255, 0, 0),
    "Hijau": (0, 255, 0),
    "Biru": (0, 0, 255)
}
```

```python
# menggunakan rumus euclidean
def find_closest_color_index(pixel):
    closest_index = None
    min_distance = float('inf')

    for index in palette:
        r1, g1, b1 = pixel
        r2, g2, b2 = palette[index]

        distance = math.sqrt((r1 - r2)**2 + (g1 - g2)**2 + (b1 - b2)**2) # rumus euclidean

        if distance < min_distance:
            min_distance = distance
            closest_index = index

    return closest_index
```

```python
# menampilkan gambar
img = Image.open('tulip3.jpg')  
img = img.convert('RGB')  
width, height = img.size
```

```python
plt.imshow(img)
plt.axis('on')
plt.title('Gambar Asli')
plt.show()
```

```python
# hasil matriks
matrix = []

for y in range(height):
    row = []
    for x in range(width):
        pixel = img.getpixel((x, y))
        index = find_closest_color_index(pixel)
        row.append(index)
    matrix.append(row)
```

```python
# membuat gambar baru dari hasil
new_image = Image.new('RGB', (width, height))

for y in range(height):
    for x in range(width):
        color = palette[matrix[y][x]]
        new_image.putpixel((x, y), color)
```

```python
# menampilkan gambar hasil menggunakan matplotlib
plt.imshow(new_image)
plt.axis('on')  
plt.title('Gambar Baru Hasil Konversi Warna')
plt.show()
```
